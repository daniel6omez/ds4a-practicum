```python
%%time
import pandas as pd
import glob

path = r'annualData2014-2019/' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    temp_df = pd.read_csv(filename, index_col=None, header=0)
    li.append(temp_df)

df = pd.concat(li, axis=0, ignore_index=True)
df.head(20)
```
